```julia
using Plots; plotlyjs()
x = -1.7:0.1:1.8
y = -2:0.1:11.8
f(x1,x2) = begin
    x = [x1,x2]
    2*x[1]^2 - x[1]*x[2] + x[2]^2 - 3*x[1] + exp(2*x[1]+x[2])
end
surface(x,y,f)
```

```julia
using Plots; plotlyjs()
x = -2:0.1:2
y = -2:0.1:2
f(x1,x2) = begin
    x = [x1,x2]
    sqrt(abs(x[1]) + abs(x[2]))
end
surface(x,y,f)
```
