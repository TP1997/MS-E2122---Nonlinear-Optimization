```julia
using Plots; plotlyjs()
x = -1.5:0.1:1.5
y = -1.5:0.1:1.5
f(x1, x2) = x1^3-x1+x2^3-x2
#plot(x,y,f,st=:surface)
surface(x,y,f)
```

```julia
using Plots; plotlyjs()
x = -1.5:0.1:1.5
y = -1.5:0.1:1.5
f(x) = x[1]^3-x[1]+x[2]^3-x[2]
#plot(x,y,f,st=:surface)
surface(x,y,f)
```

```julia
using ForwardDiff
H(f,x) = ForwardDiff.hessian(f,x)
```

```julia
using LinearAlgebra
cp = 1/sqrt(3)
ev1=eigvals(H(f,[cp,cp]))
println(ev1)
ev2=eigvals(H(f,[cp,-cp]))
println(ev2)
ev3=eigvals(H(f,[-cp,cp]))
println(ev3)
ev4=eigvals(H(f,[-cp,-cp]))
println(ev4)
```
